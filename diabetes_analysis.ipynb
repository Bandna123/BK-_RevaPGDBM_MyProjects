```python
#Bin variables
import numpy as np
diabetes['GlucoseBin'] = np.where(diabetes['Glucose'] <=140,'<=140', (np.where(diabetes['Glucose'] >140 ,'>140', '0')))
diabetes.head()
```

```python
#Bin variables
import numpy as np
diabetes['BMIBin'] = np.where(diabetes['BMI'] <=25,'<=25', (np.where(diabetes['Glucose'] >25 ,'>25', '0')))
diabetes.head()
```

```python
pd.pivot_table(train, values="Glucose", index=["GlucoseBin"], columns="Outcome", aggfunc = "count",fill_value=0)
```
